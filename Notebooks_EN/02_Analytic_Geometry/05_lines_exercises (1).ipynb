```python
# The line passes through points A(1,2) and B(3,4). Find the equation of the line

import numpy as np
import matplotlib.pyplot as plt

A = np.array([1, 2])
B = np.array([3, 4])

m = (B[1] - A[1]) / (B[0] - A[0])

b = A[1] - m * A[0]

equation = f"y = {m}x + {b}"

print(f"The equation of the line is: y = {m:.2f}x + {b:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equation
def funcion_lineal(x):
    return x + 1

# x values
x = np.linspace(-10, 10, 400)

# calculate y values
y = funcion_lineal(x)

# Represent
plt.plot(x, y, label=r'$y = x + 1$', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphic of $y = x + 1$')
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
print(f"The equation of the line is: y = x + 1")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equation
def funcion_lineal(x):
    return 2 * x

x = np.linspace(-10, 10, 400)

y = funcion_lineal(x)

plt.plot(x, y, label=r'$y = 2x$', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphic of $y = 2x$')
plt.legend()
plt.grid(True)

plt.show()
print(f"The equation of the line is: y = 2x")
```

```python
# The line passes through point A(1,2) and is perpendicular to the line y=2x+3. Find the equation of the line

x1, y1 = 1, 2
m = 2
m_perpendicular = -1 / m

b = y1 - m_perpendicular * x1

equation = f"y = {m_perpendicular}x + {b}"

print("The equation of the line is:", equation)
```

```python
# We have two lines y=2x+3 and y=3x+2. Find the intersection point of these lines and calculate the angle between them.

import math

m1 = 2
m2 = 3

n1 = 3
n2 = 2

x = (n2 - n1) / (m1 - m2)

y = m1 * x + n1

rad = math.atan(abs((m1 - m2) / (1 + m1 * m2)))

deg = math.degrees(rad)

print(f"Intersection Point: ({x}, {y})")
print(f"Angle between the lines: {deg:.2f} degrees")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Lines parameters
m1, n1 = 2, 3  # y = 2x + 3
m2, n2 = 3, 2  # y = 3x + 2

# Intersection point
x_inter = (n2 - n1) / (m1 - m2)
y_inter = m1 * x_inter + n1

# Angle calculation for the two lines
rad = math.atan(abs((m2 - m1) / (1 + m1 * m2)))
deg = math.degrees(rad)

# Rango de valores para graficar las rectas
x_vals = np.linspace(-5, 5, 400)

# Ecuaciones de las rectas
y1_vals = m1 * x_vals + n1  # y = m1*x + n1
y2_vals = m2 * x_vals + n2  # y = m2*x + n2

# Crear la figura y el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1_vals, label=f'$y = {m1}x + {n1}$', color='blue')
plt.plot(x_vals, y2_vals, label=f'$y = {m2}x + {n2}$', color='green')

# Resaltar el punto de intersección
plt.scatter(x_inter, y_inter, color='red', zorder=5, label=f'Intersection: ({x_inter:.2f}, {y_inter:.2f})')

# Añadir detalles al gráfico
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title(f'Intersection and Angle: {deg:.2f}°', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()

# Imprimir los resultados en consola
print(f"Intersection Point: ({x_inter:.2f}, {y_inter:.2f})")
print(f"Angle between the lines: {deg:.2f} degrees")
```

```python
# We have the line y=2x+3 and point A(1,2). Find the distance from point A to the line.

import math

A, B, C = 2, -1, 3
x, y = 1, 2

distance = abs(A*x + B*y + C) / math.sqrt(A**2 + B**2)

print(f"The distance from point A(1, 2) to the line y = 2x + 3 is: {distance:.4f}")
```
